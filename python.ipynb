## Aula 3 - Tabuada e Divis√£o ü•∫ üëâ üëà

```python
def tabuada(numero):
    for contador in range(1, 11):  # Gera n√∫meros de 1 a 10
        print(f"{numero} x {contador} = {numero * contador}")
numero = int(input("Digite um n√∫mero para ver sua tabuada: "))
tabuada(numero)
```

```python
def divisao(numero):
    for contador in range(1, 11):  # Gera n√∫meros de 1 a 10
        resultado = numero / contador
        if resultado.is_integer():  # Verifica se o resultado √© inteiro
            print(f"{numero} / {contador} = {int(resultado)}")
        else:
            print(f"{numero} / {contador} = {resultado:.2f}")  # Mostra at√© 2 casas decimais

numero = int(input("Digite um n√∫mero para ver sua divis√£o: "))
divisao(numero)
```

##  Aula 4 - imc ü§´ üßèüèª‚Äç‚ôÇÔ∏è üóø

```python
peso = float(input("Digite seu peso (em kg): "))
altura = float(input("Digite sua altura (em metros): "))
imc = peso / (altura ** 2)
print(f"\nSeu IMC √©: {imc:.2f}")
if imc < 18.5:
    print("Voc√™ est√° abaixo do peso.")
elif 18.5 <= imc < 25:
    print("Voc√™ est√° no peso ideal.")
elif 25 <= imc < 30:
    print("Voc√™ est√° com sobrepeso.")
else:
    print("Voc√™ est√° acima do peso.")
```

## Aula 6 - Listas üõ©Ô∏è üè¢ üè¢
